```python
striped_news_data = news_data.drop(['source', 'subject', 'text', 'title', 'url'], axis = 1).reset_index(drop=True)
striped_news_data.columns
```

```python
striped_news_data['date'] = pd.to_datetime(striped_news_data['date'], format='%d-%m-%Y')
```

```python
striped_news_data['sentiment'] = striped_news_data['sentiment'].map(lambda x: 1 if x > 0 else -1 if x < 0 else 0)
```
